```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Upload and Load Dataset

uploaded = files.upload()

# Load dataset
df = pd.read_csv("Advertising.csv")

print("===== First 5 Rows of Dataset =====")
print(df.head())
```

```python
# Data Preparation


# Features (X) and Target (y)
X = df.drop("sales", axis=1)
y = df["sales"]

print("\n===== Features =====")
print(X.head())

print("\n===== Target =====")
print(y.head())


# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=101
)

print("\n===== Train/Test Split =====")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

```

```python
# Linear Regression Model

# Create Model
linear_model = LinearRegression()

# Train Model
linear_model.fit(X_train, y_train)

# Predictions
linear_predictions = linear_model.predict(X_test)

# Evaluation Metrics
linear_MAE = mean_absolute_error(y_test, linear_predictions)
linear_MSE = mean_squared_error(y_test, linear_predictions)
linear_RMSE = np.sqrt(linear_MSE)

print("\n===== Linear Regression Results =====")
print("MAE:", linear_MAE)
print("MSE:", linear_MSE)
print("RMSE:", linear_RMSE)
```

```python
#Polynomial Regression Model

# Create Polynomial Features (degree = 2)
poly_converter = PolynomialFeatures(
    degree=2,
    include_bias=False
)

# Fit ONLY on training data (avoids data leakage)
poly_converter.fit(X_train)

# Transform train and test data separately
X_train_poly = poly_converter.transform(X_train)
X_test_poly = poly_converter.transform(X_test)

# Create Model
poly_model = LinearRegression()

# Train Model
poly_model.fit(X_train_poly, y_train)

# Predictions
poly_predictions = poly_model.predict(X_test_poly)

# Evaluation Metrics
poly_MAE = mean_absolute_error(y_test, poly_predictions)
poly_MSE = mean_squared_error(y_test, poly_predictions)
poly_RMSE = np.sqrt(poly_MSE)

print("\n===== Polynomial Regression Results =====")
print("MAE:", poly_MAE)
print("MSE:", poly_MSE)
print("RMSE:", poly_RMSE)
```

```python
# =========================================
# 6. Model Comparison
# =========================================

comparison = pd.DataFrame({
    "Model": ["Linear Regression", "Polynomial Regression"],
    "MAE": [linear_MAE, poly_MAE],
    "MSE": [linear_MSE, poly_MSE],
    "RMSE": [linear_RMSE, poly_RMSE]
})

print("\n===== Model Comparison =====")
print(comparison)


# Identify Better Model
if poly_RMSE < linear_RMSE:
    print("\nPolynomial Regression performs better because it has lower RMSE.")
else:
    print("\nLinear Regression performs better because it has lower RMSE.")
```

```python
# Coefficients Interpretation


print("\n===== Linear Regression Coefficients =====")

coef_df = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": linear_model.coef_
})

print(coef_df)

print("\nIntercept:", linear_model.intercept_)


# Simple Interpretation
print("\n===== Interpretation =====")
print("Positive coefficient → increasing this feature increases sales.")
print("Negative coefficient → increasing this feature decreases sales.")
print("Larger absolute value → stronger effect on prediction.")
```

The results show that Polynomial Regression performed better than Linear Regression on the Advertising dataset because it achieved lower error values in MAE, MSE, and RMSE. Linear Regression had an RMSE of 1.516, while Polynomial Regression reduced it significantly to 0.664, indicating more accurate predictions. This suggests that the relationship between advertising features and sales is not perfectly linear, and Polynomial Regression captures the pattern more effectively.
